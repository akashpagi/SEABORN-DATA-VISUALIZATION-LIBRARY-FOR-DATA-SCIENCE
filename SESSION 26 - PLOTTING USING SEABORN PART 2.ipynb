**Categorical Scatter Plot**
- Perform bivariate analysis

**Stripplot**
- Draw a categorical scatterplot using jitter to reduce overplotting.

- **Syntax: sns.stripplot(data, x, y, hue,  jitter=True,orient, color, size=5, linewidth=0,legend='auto')**

````python
# stripplot - jitter, hue
# axes level function
sns.stripplot(tips, x='day', y='total_bill', jitter=False, hue='day')
````

**Swarmplot**
- Draw a categorical scatterplot with points adjusted to be non-overlapping.

- **Syntax: sns.swarmplot(data, x, y, hue,  jitter=True, color, size=5, linewidth=0,legend='auto')**

````python
# axes-level function
sns.swarmplot(tips, x='day', y='total_bill', hue='sex', linewidth=1)
````

**catplot**
- Figure-level interface for drawing categorical plots
- **Syntax: sns.catplot(data, x, y, kind='strip' hue,  jitter=True, col, row, color, size=5, linewidth=0,legend='auto')**

````python
# axes level function
sns.catplot(tips, x='day', y='total_bill', kind='strip', 
            jitter=0.3, hue='sex')
````

````python
# figure-level function
sns.catplot(tips, x='day', y='total_bill', kind='swarm', hue='sex')
````

**Boxplot**
- Draw a box plot to show distributions with respect to categories.
- A boxplot is a standardized way of displaying the distribution of data based on a five number summary (â€œminimumâ€, first quartile [Q1], median, third quartile [Q3] and â€œmaximumâ€).
- It can tell you about your outliers and what their values are.
- Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.
 
- **Syntax: sns.boxplot(data, x, y, hue,linecolor='auto',color, fill=True,saturation=0.75, dodge='auto',width=0.8, gap=0,whis=1.5,linewidth=0,legend='auto')**

````python
#boxplot - data, x, y, hue, linewidth=2
# per day sex ratio
sns.boxplot(tips, x='day', y='total_bill', hue='sex',
           linewidth=2, fill=False, dodge=True)
````

#### **Regression plots**
- A regression plot in Seaborn is a visualization that combines a scatter plot with a fitted regression line, helping you understand the linear relationship between two variables in your data. It's useful for exploring and visualizing trends, especially in the context of building predictive models.
- **regplot**
- **lmplot**
- **residplot**
  
- **Note:** In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

**Regplot**
- Plot data and a linear regression model fit.
- **Syntax: sns.regplot(data, x, y, x_estimator, x_bins, x_ci='ci', scatter=True, fit_reg=True, order, truncate=True, dropna=True, x_jitter, y_jitter, label, color, marker='o')**

````python
# regplot > axes level function
sns.regplot(tips, x='total_bill', y='tip')
````

**lmplot**
- Plot data and regression model fits across a FacetGrid.
- Combine **regplot()** and **FacetGrid** to plot multiple linear relationships in a dataset.
- **Syntax: sns.lmplot(data, x, y, hue, col, row, palette, col_wrap, height=5, aspect=1, markers='o', sharex, sharey, legend=True, x_estimator, x_bins, x_ci='ci', scatter=True, fit_reg=True )**

````python
# lmplot > figure level function
sns.lmplot(tips, x='total_bill', y='tip', hue='sex')
````

**Residplot**
- Plot the residuals of a linear regression.
- **Synatx: sns.residplot(data, x, y, x_partial, y_partial, lowess=False, order=1, robust=False, dropna=True, label, color)**

````python
# residplot
sns.residplot(tips, x='total_bill', y='tip')
````

#### **Multi-plot grids**

**Facet grids**
- Multi-plot grid for plotting conditional relationships.
- **Syntax: sns.FacetGrid(data, row, col, hue, col_wrap, sharex=True, sharey=True, height=3,margin_titles=False)**

**FacetGrid.map(func, `*args`, `**kwargs`)**
- Apply a plotting function to each facetâ€™s subset of the data.

````python
# factgrid
fg = sns.FacetGrid(tips, col="smoker", row="sex", hue='smoker')
fg.map(sns.scatterplot, "total_bill", "tip")
fg.add_legend()
````

- To draw a plot on every facet, pass a function and the name of one or more columns in the dataframe to FacetGrid.map()

````python
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
````

#### **Pair grids**

**pairplot**
- Plot pairwise relationships in a dataset.
```python
Syntax: sns.pairplot(data, hue,
                       diag_kind:{â€˜autoâ€™, â€˜histâ€™, â€˜kdeâ€™, None},
                       kind:{â€˜scatterâ€™, â€˜kdeâ€™, â€˜histâ€™, â€˜regâ€™},
                       x_vars, y_vars, markers=["o", "s", "D"], 
                       corner=False, fill=True,
                       height=2.5 (in Inches),  dropna=False)

````python
# pairplot > hue, markers, fill, kind parameters
# It is symmetircal plots, 
sns.pairplot(iris, hue="species", markers=["o", "s", "D"],
             diag_kind='kde', corner=False, height=2.5)
````

**PairGrid**
- Subplot grid for plotting pairwise relationships in a dataset.
```python
Syntax: sns.PairGrid(data, hue, x_vars, y_vars, 
                     corner=False, height=2.5, dropna=False)*

**pairplot vs PairGrid**
- In Pairplot there is restriction to drw plots and in PairGrid no restriction for plots

- Passing separate functions to **PairGrid.map_diag()** and **PairGrid.map_offdiag()** will show each variableâ€™s marginal distribution on the diagonal:

- **PairGrid.map_diag(`func`, `**kwargs`)**
    - Plot with a univariate function on each diagonal subplot.

- **PairGrid.map_offdiag(`func`, `**kwargs`)**
    - Plot with a bivariate function on the off-diagonal subplots.

````python
# PairGrid
pg = sns.PairGrid(iris, hue='species')
pg.map(sns.scatterplot)
````

````python
# map_diag & map_offdiag functions
pg = sns.PairGrid(iris,  hue='species')
pg.map_diag(sns.boxplot)
pg.map_offdiag(sns.histplot)
````

````python
# fexibility
pg = sns.PairGrid(iris, hue='species', vars=['sepal_width','petal_width'])
pg.map_diag(sns.histplot)
pg.map_upper(sns.kdeplot)
pg.map_lower(sns.scatterplot)
````

#### **Join grids**

**jointplot**
- Draw a plot of two variables with bivariate and univariate graphs.
```python
Syntax: sns.jointplot(data, x, y, hue, kind='scatter', 
                      height=6, ratio=5, space=0.2, 
                      dropna=False, xlim=None, ylim=None, 
                      color=None, marginal_ticks=False)

````python
# jointplot
sns.jointplot(data=tips,x='total_bill',y='tip',
              kind='kde',hue='sex', height=3, space=0.5,
              marginal_ticks=True)
````

**JointGrid**
- Grid for drawing a bivariate plot with marginal univariate plots.

```python
Syntax: sns.JointGrid(data, x, y, hue, kind='scatter', 
                      height=6, ratio=5, space=0.2, 
                      dropna=False, xlim=None, ylim=None, 
                      color=None, marginal_ticks=False) 

**JointGrid.plot(`joint_func`, m`arginal_func`, `**kwargs`)**
- Draw the plot by passing functions for joint and marginal axes.

````python
# JointGrid &
jg = sns.JointGrid(tips, x='total_bill', y='tip', hue='sex')
jg.plot(sns.scatterplot, sns.boxplot)
````
